```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from datetime import date
```

```python
main_config = {
    "hovermode": "x unified",
    "legend": {"yanchor":"top",
                "y":0.9,
                "xanchor":"left",
                "x":0.1,
                "title": {"text": None},
                "font" :{"color":"white"},
                "bgcolor": "rgba(0,0,0,0.5)"},
    "margin": {"l":0, "r":0, "t":20, "b":0}
}
```

```python
data = date.today()
data_ano = data.year

dfM = pd.read_csv("sib_202307_AC.csv", sep = ";", encoding="latin-1")
dfM["Ano_Atual"] = data_ano
```

```python
dfM2 = dfM[["DT_NASCIMENTO", "Ano_Atual"]]
dfM2["Idade"] = dfM["Ano_Atual"] - dfM["DT_NASCIMENTO"]
dfM2
```

```python
dfM2["Idade Média"] = round(dfM2["Idade"].mean())
dfM2
```

```python
fig = go.Figure()
fig.add_trace(

   go.Indicator(mode="number",
                title = {"text": "<span style = 'font-size:100%> Idade média </span><br>"},
                value = dfM2['Idade Média'].iloc[0])

)
```

```python
dfM2
```

```python
P60 = (len(dfM2[ dfM2["Idade"] >= 60]) / len(dfM2)) * 100
P60
```

```python
fig = go.Figure()
fig.add_trace(

   go.Indicator(mode="number",
                title = {"text": "<span style = 'font-size:100%> % Idosos </span><br>"},
                value = P60)

)
```

```python
P14 = len(dfM2[ dfM2["Idade"] <= 14])
P14
```

```python
Indice_Envelhecimento = (P60 / P14) * 100
Indice_Envelhecimento
```

```python
fig = go.Figure()
fig.add_trace(

   go.Indicator(mode="number",
                title = {"text": "<span style = 'font-size:100%> Índice Envelhecimento </span><br>"},
                value = Indice_Envelhecimento)

)
```

```python
Razao_Dependencia = (P14 / P60) * 100
Razao_Dependencia
```

```python
fig = go.Figure()
fig.add_trace(

   go.Indicator(mode="number",
                title = {"text": "<span style = 'font-size:100%> Razão de Dependência </span><br>"},
                value = Razao_Dependencia)

)
```
